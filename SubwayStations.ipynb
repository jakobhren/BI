```python
import kagglehub
import csv
import requests
import time
import os
import pandas as pd
from dotenv import load_dotenv
import io
```

```python
load_dotenv()
```

```python
path = kagglehub.dataset_download("new-york-state/nys-nyc-transit-subway-entrance-and-exit-data")

# Find the CSV file in the downloaded dataset folder
csv_files = [f for f in os.listdir(path) if f.endswith(".csv")]
csv_file = os.path.join(path, csv_files[0])

subway_stations = pd.read_csv(csv_file)

print("Loaded rows:", len(subway_stations))
subway_stations.head()
```

```python
cols_keep = ["Station Name", "Line", "Station Longitude", "Station Latitude", "Route1", "Route2", "Route3", "Route4", "Route5", "Route6", "Route7", "Route8", "Route9", "Route10", "Route11"]
```

```python
subway_stations = subway_stations[cols_keep]
```

```python
subway_stations.head()
```

```python
entrance_count = subway_stations.groupby(['Station Name', 'Line']).size().reset_index(name='EntranceCount')
subway_stations = subway_stations.merge(entrance_count, on=['Station Name', 'Line'], how='left')
```

```python
subway_stations.head(20)
```

```python
unique_count = subway_stations.groupby(['Station Name', 'Line']).size().reset_index(name='Count')
print(f"Number of unique Station Name and Line combinations: {len(unique_count)}")
unique_count.head(5)
```

```python
subway_stations = subway_stations.drop_duplicates(subset=['Station Name', 'Line'], keep='first')
```

```python
subway_stations.head()
```

```python
def count_routes(row):
    routes = [row['Route1'], row['Route2'], row['Route3'], row['Route4'], 
              row['Route5'], row['Route6'], row['Route7'], row['Route8'], 
              row['Route9'], row['Route10'], row['Route11']]
    return sum(1 for route in routes if pd.notna(route))

subway_stations['RouteCount'] = subway_stations.apply(count_routes, axis=1)
```

```python
subway_stations.head()
```

```python
subway_stations = subway_stations.drop(columns=[f'Route{i}' for i in range(1, 12)])
```
