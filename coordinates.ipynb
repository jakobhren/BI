```python
import requests
import time
import pandas as pd
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
USER_AGENT = os.getenv("USER_AGENT")

def geocode_neighborhood(name):
    url = "https://nominatim.openstreetmap.org/search"
    params = {
        "q": f"{name}, New York City",
        "format": "json",
        "limit": 1
    }
    headers = {"User-Agent": USER_AGENT}

    
    r = requests.get(url, params=params, headers=headers)
    data = r.json()
    
    if len(data) == 0:
        return None, None
    return float(data[0]["lat"]), float(data[0]["lon"])

neighborhoods = [
    'Upper West Side', 'Upper East Side', 'East New York', 'Bedford-Stuyvesant',
    'Washington Heights', 'Flushing', 'Williamsburg', 'Crown Heights', 'Harlem',
    'Astoria', 'Borough Park', 'Sheepshead Bay', 'Bushwick', 'East Flatbush',
    'Concourse', 'Gravesend', 'Elmhurst', 'Midwood', 'Canarsie', 'Jackson Heights',
    'Sunset Park', 'Soundview', 'East Harlem', 'Jamaica', 'Bensonhurst',
    'Bay Ridge', 'Far Rockaway', 'Brownsville', 'Forest Hills', 'Lower East Side',
    'Corona', 'East Village', 'Ridgewood', 'Park Slope', 'Flatbush', 'Woodhaven',
    'Woodside', 'Queens Village', 'Fordham', 'St. Albans', 'Hamilton Heights',
    'Wakefield', "Hell's Kitchen", 'Tremont', 'Mott Haven', 'Maspeth', 'Chelsea',
    'Bayside', 'Kensington', 'East Tremont', 'Richmond Hill', 'Kingsbridge',
    'Riverdale', 'Williamsbridge', 'Springfield Gardens', 'Rego Park',
    'Coney Island', 'Kew Gardens Hills', 'Parkchester', 'Fresh Meadows',
    'Morris Heights', 'Whitestone', 'Inwood', 'Dyker Heights', 'Highbridge',
    'South Ozone Park', 'Greenpoint', 'Norwood', 'Middle Village', 'Glendale',
    'Throggs Neck', 'Bath Beach', 'Greenwich Village', 'Bedford Park',
    'Castle Hill', 'Brighton Beach', 'Fort Greene', 'Prospect Lefferts Gardens',
    'Financial District', 'Morningside Heights', 'Sunnyside', 'Howard Beach',
    'Laurelton', 'Murray Hill', 'College Point', 'West Village', 'Hollis',
    'Briarwood', 'Kew Gardens', 'Oakland Gardens', 'Clinton Hill',
    "Mariner's Harbor", 'Rosedale', 'Prospect Park South', 'Melrose', 'Belmont',
    'Eltingville', 'Flatlands', 'Downtown', 'New Springville', 'Marine Park',
    'Brooklyn Heights', 'Great Kills', 'Bellerose', 'Ozone Park',
    'Jamaica Estates', 'Cambria Heights', 'Prospect Heights', 'Tribeca',
    'Flatiron District', 'Turtle Bay', 'Sutton Place', 'Port Richmond',
    'Morris Park', 'Jamaica Hills', 'West Brighton', 'Gramercy',
    'Carroll Gardens', 'Greenwood', 'East Elmhurst', 'Longwood',
    'Windsor Terrace', 'Van Nest', 'Boerum Hill', 'Pelham Gardens',
    'Battery Park', 'Rossville', 'Huguenot', 'Floral park', 'Graniteville',
    'Pelham Bay', 'Wingate', 'Midtown', 'Bergen Beach', 'Glen Oaks',
    'Rockaway Park', 'Douglaston', 'Dongan Hills', 'Annadale', 'Tottenville',
    'Castleton Corners', 'Midland Beach', 'Cobble Hill', 'Roosevelt Island',
    'SoHo', 'South Beach', 'St. George', 'Hudson Yards', 'Red Hook', 'New Dorp',
    'Richmond Town', 'Baychester', 'Gowanus', 'Clifton', 'Tompkinsville',
    'Park Hill', "Prince's Bay", 'Eastchester', 'DUMBO', 'Shore Acres',
    'Westerleigh', 'Meiers Corners', 'Country Club', 'Todt Hill', 'Woodlawn',
    'Belle Harbor', 'Bay Terrace', 'New Brighton', 'Woodrow', 'Mill Basin',
    'Oakwood', 'Arden Heights', 'Bulls Head', 'Grymes Hill', 'Emerson Hill',
    'Rosebank', 'Sunnyside', 'Manhattan Beach', 'New Dorp Beach', 'Grant City',
    'Columbia Street Waterfront District', 'Elm Park', 'City Island',
    'Gerritsen Beach', 'Willowbrook', 'Stapleton', 'Little Italy',
    'Pleasant Plains', 'Arrochar', 'Silver Lake', 'Richmond Valley', 'Greenridge',
    'Charleston', 'Travis', 'Grasmere', 'Broad Channel', 'Lighthouse Hill',
    'Neponsit'
]


results = []

for n in neighborhoods:
    lat, lon = geocode_neighborhood(n)
    print(f"{n} → {lat}, {lon}")
    results.append({"neighborhood": n, "latitude": lat, "longitude": lon})
    time.sleep(1)  # required: avoid rate-limiting

df = pd.DataFrame(results)
print(df)
```
